### Synopsis

This analysis explores the closure of San Francisco businesses in response to the COVID-19 Pandemic. We explore questions about the rate of closures over the last year, and which industries and neighborhoods seemed to be hit the hardest. To aid in this analysis, we use registered business data provided by San Francisco at [data.sfgov.org](https://data.sfgov.org/Economy-and-Community/Registered-Business-Locations-San-Francisco/g8m3-pdis). This dataset is updated weekly and as of writing, contains records of 268,000 businesses registered with the City and County of San Francisco. 

For our analysis, primarily focus on the following data fields for our analysis:
- `Location End Date`
- `NAICS Code Description`
- `Neighborhoods - Analysis Boundaries`
- `Unique ID`

We use a few other related fields to help us clean and validate the data, but we'll discuss those throughout the report. For our analysis, we restrict the data to only those business locations that were in San Francisco and were open prior to March 1st, 2020. Some data vetting was needed to get rid of unusable records, in which the dates did not make logical sense (i.e. the location ended before it start, the business ended before it started, the business ended before the physical location ended, etc). After filtering based on this criteria, we perform an analysis on 107,703 records.

Key Findings:
1) Per our data, 5,084 business locations have shuttered in San Francisco since March 1st and as of writing. This is 4.7% of those that were open as of that same date. As a benchmark, the SF Chronicle reported that 2,065 business had permanently closed as of July 10th, 2020 per Yelp's analysis of business listings on their site. Our analysis finds that 2,368 had closed by that same date.
2) Daily closures as reported int eh data seem to spike on the first and last days of each month. This could just be due to data reporting, but could also be due to when business decisions are made. Exceptions to this are spikes in closures from March 15th-March 17th, 2020. These spikes may have been related to the looming shelter-in-place order that was ultimately enacted March 16th.
3) Roughly 16% of all closures recorded over the last year occured in March of 2020. This % declined a bit over the following months, however we see a sudden spike again in December. This spike may be due in part to the year ending, but it also coincides with a new wave of COVID-19 and a renewed regional stay-at-home-order issued on December 6th, 2020.
4) The "Certain Services" sector had the highest closure rate of any sector, with 7.4%. These codes refer to several common maintenence and repair businesses, such as automative, computer and electronic repair, household maintenance services, personal care services (i.e. hair and nail salons) and more. Retail and Food Services hover at around a 6% closure rate, whereas manufacturing, construction and real estate ranged from 2-3%.
5) Japantown had the highest closure rate of any neighborhood, with roughly 62 companies per thousand shutting down over the course of the pandemic. This is notably higher than the 47 companies per thousand that shut down overall. Glen Park, Visitacion Island, and Treasure Island faired better, with roughly 20-30 companies per thousand shutting down.

Things to Note:

### Cleaning the Data

Here, we just want to tidy up the column names and find valid records based on the physical business location and business/location start and end dates. We'll describe the criteria for this as we go, so that we can also see how many records are plagued by each issue we uncover.

#### Column Names
First, let's get the column names into a tidy format.

```python
# Transforms the column names into snakecase.
raw_data.columns = (raw_data.
                    columns.
                    str.lower().
                    str.replace("-", "").
                    str.replace("\s+", "_", regex=True))

raw_data.head()
```

There! Much prettier and easier to work with. Next, let's try to classify each record into whether the provided business/location is open, closed, and whether or not its dates make sense.

#### Invalid Dates

Here, we can make a new column that designates the status of the record: open, closed, or invalid:
- Records with `invalid` status are those in which:
    - `location_start_date` precedes the `business_start_date` or
    - `business_end_date` precedes the `location_end_date` or
    - `location_end_date` precedes the `location_start_date` or
    - `business_end_date` precedes the `business_start_date`
- Records with `open` status are those in which both the `business_end_date` and `location_end_date` are null.
- Records with `closed` status are those in which `location_end_date` contains some date value, indicating closure.

```python
# Filters the data based on the conditions described above. First classified invalid records
# then classified open records, and lastly, closed records.
raw_data = (raw_data.
            assign(
                status = np.where(
                    (raw_data['location_start_date'] < raw_data['business_start_date']) |
                    (raw_data['business_end_date'] < raw_data['location_end_date']) |
                    (raw_data['location_end_date'] < raw_data['location_start_date']) |
                    (raw_data['business_end_date'] < raw_data['business_start_date']) |
                    (pd.notnull(raw_data['business_end_date']) & pd.isnull(raw_data['location_end_date'])), 
                    "invalid", 
                    np.where(
                        pd.isnull(raw_data['business_end_date']) &
                        pd.isnull(raw_data['location_end_date']),
                        "open",
                        "closed"
                    ))))
```

Now that we've classified each record, let's see how many are still good to use and how many locations are open or closed.

```python
(raw_data.
 pipe(count_vals, "status", "uniqueid"))
```

Roughly 1% of all records had issues with their dates and were invalid. Let's go ahead and remove them from our data.

```python
valid_records = (raw_data.
                 query('status != "invalid"'))
```

#### Locations Outside SF

Next, let's check if any records are not located within San Francisco. For our analysis, we're only interested in those business locations that are physically in San Francisco (as opposed to San Jose, Oakland, etc). We'll do this by checking to see if the record's business location `city` or `mail_city` are San Francisco.

```python
(valid_records.
 assign(
     within_sf = np.where(
         (valid_records['city'] == "San Francisco") | 
         (valid_records['mail_city'] == "San Francisco"), 
         "sf", 
         "not_sf"
 )).
 pipe(count_vals, "within_sf", "uniqueid"))
```

46,000 feels like a fairly large number of records to be outside San Francisco. Let's see if we're missing anything. Are any records listed as `not_sf` tied to a San Francisco neighborhood? We can check this by filtering the records to those `not_sf` and then counting the values still in `neighborhoods_analysis_boundaries`.

```python
(valid_records.
 assign(
     within_sf = np.where(
         (valid_records['city'] == "San Francisco") | 
         (valid_records['mail_city'] == "San Francisco"), 
         "sf", 
         "not_sf"
 )).
 query('within_sf == "not_sf"').
 pipe(count_vals, "neighborhoods_analysis_boundaries", "uniqueid"))
```

These records could actually be in San Francisco, but perhaps a spelling error is why we didn't catch them before. Let's take a closer look at those in `South of Market` as an example.

```python
(valid_records.
 assign(
     within_sf = np.where(
         (valid_records['city'] == "San Francisco") | 
         (valid_records['mail_city'] == "San Francisco"), 
         "sf", 
         "not_sf"
 )).
 query(
     'within_sf == "not_sf" & \
     neighborhoods_analysis_boundaries == "South of Market"'
 ).
 head())
```

It's clear that these records should be included in our analysis, but due to spelling errors in the `city` and `mail_city` columns, they were excluded. Let's modify our criteria and see how many records are actually in San Francisco as opposed to elsewhere. Here, we'll assume that if the record was assigned to a San Francisco neighborhood, it is in the city. Again, we'll use the `neighborhoods_analysis_boundaries` field to test this.

```python
(valid_records.
 assign(
     within_sf = np.where(
         (valid_records['city'] == "San Francisco") | 
         (valid_records['mail_city'] == "San Francisco") |
         (pd.notnull(valid_records['neighborhoods_analysis_boundaries'])), 
         "sf", 
         "not_sf"
 )).
 pipe(count_vals, "within_sf", "uniqueid"))
```

It's a small change, but we've managed to include about 1,700 records that we otherwise might've missed. Let's do one last check to see what cities compromise the records we've labeled as `not_sf`.

```python
(valid_records.
 assign(
     within_sf = np.where(
         (valid_records['city'] == "San Francisco") | 
         (valid_records['mail_city'] == "San Francisco") |
         (pd.notnull(valid_records['neighborhoods_analysis_boundaries'])), 
         "sf", 
         "not_sf"
 )).
 query('within_sf == "not_sf"').
 pipe(count_vals, "city", "uniqueid").
 head(10))
```

We notice that South San Francisco is included here as `S San Francisco`, however given that it is not part of San Francisco county, we feel good about excluding it. Let's go ahead and filter down our `valid_records` to just those within San Francisco! We should be left with about 220K records.

```python
valid_records = (valid_records.
                 assign(
                     within_sf = np.where(
                         (valid_records['city'] == "San Francisco") | 
                         (valid_records['mail_city'] == "San Francisco") |
                         (pd.notnull(valid_records['neighborhoods_analysis_boundaries'])), 
                         "sf", 
                         "not_sf"
                 )).
                 query('within_sf == "sf"'))

valid_records.uniqueid.count()
```

#### Filtering to Relevant Dates

Here, we'll filter down our data to just those business locations which were open as of March 1st, 2020, and are either open as of writing, or closed after the Pandemic start date we listed above. By doing this we focus our sample on just the business locations that were actually open as of the Pandemic start, rather than those that opened afterwards. That could be a separate analysis later on. We'll do this using the `location_start_date`, `location_end_date`, and the recently created `status` field.

```python
sf_records = (valid_records.
              query(
                  'location_start_date <= @pandemic_start_date & \
                  (status == "open" | \
                  location_end_date >= @pandemic_start_date)'))

sf_records.uniqueid.count()
```

This gives us 107,703 records with which to work. This is a fairly drastic decrease from ~220,000, but it makes sense given we're focusing on a single year of a dataset that spans over a century.

```python
sf_records.head()
```

### Analyzing COVID's Impact

#### Closures Over Time

Here we start to dive into the qualified data. At a high level, we'll begin by exploring how many businesses cumulatively closed each day and week following March 1st, 2020.

```python
closures_by_day = (sf_records.
                    groupby("location_end_date")['uniqueid'].
                    count().
                    reset_index().
                    rename(columns = {"location_end_date": "closure_date", "uniqueid": "closures"}))

closures_by_day["cumulative_closures"] = closures_by_day["closures"].cumsum()

closures_by_day = (closures_by_day.
                    assign(pct_of_total_closures = closures_by_day['cumulative_closures'] / closures_by_day['closures'].sum()))

closures_by_day.query('closure_date in ["2020-07-10", "2021-03-08"]')
```

#### Closures by Industry

Now that we have a sense of when closures happened, we can begin to look at how COVID impacted different sectors. It's reasonable to assume that COVID had varying impact as restaurants and food services were undoubtedly hit harder than companies making hand sanitizer, for instance. Let's see how we can do this.

```python
sf_records.head()
```

The data comes with NAICS (North American industry Classification) codes. These seem like the perfect candidate. Even from our brief glimpse of the data, it's clear that not every record is classified using these codes. Let's see how many are classified and to which codes they belong.

```python
sf_records.naics_code_description.count()
```

Of the roughly 107K thousand records we're using, only 87K have associated NAICS codes. For the following analysis, we'll focus on them.*

**Note: In a more rigorous analysis, we would want to see how the characteristics of these soon-to-be-removed records differ from those we're going to keep. Should both sets of data be similar, our actions aren't particularly problematic, but should they differ, our below analysis may not be particularly representative.*

```python
sf_records.pipe(count_vals, "naics_code_description", "uniqueid")
```

We see that real estate and professional, scientific, and technical services dominate, followed by retail trade, food services, and health services. Let's now try to calculate the closure rate for each of these sectors. To do so, we'll have to tally up how many of the business location in each closed over the last year.

```python
closures_by_sector = (sf_records.
                      groupby(["naics_code_description", "status"])["uniqueid"].
                      count().
                      reset_index().
                      rename(columns={"naics_code_description": "sector"}).
                      pivot(index="sector", columns="status", values="uniqueid").
                      assign(
                          pct_closed = lambda x: x.closed / (x.closed + x.open),
                          rate_per_thousand = lambda x: x.pct_closed * 1000
                          #rounded = lambda x: x.rate_per_thousand.round(1)
                      ).
                      reset_index().
                      sort_values(by="pct_closed", ascending=False))

closures_by_sector
```

##### **Finding 4**

We see that `Certain Services` are at the top. According to the NCAIS, these codes refer to several common maintenence and repair businesses, such as automative, computer and electronic repair, household maintenance services, personal care services (i.e. hair and nail salons) and more. A more complete list can be found on the [NAICS site](https://www.naics.com/six-digit-naics/?code=81) directly.

Moving along, we see that retail and food service sectors hover at around a 6% closure rate, whereas more resilient sectors such as manufacturing, construction and real estate faired much better. We can visualize these findings to get a better sense of things.

##### Chart 4: Which Sectors in San Francisco Were Hit the Hardest?

```python
(ggplot(closures_by_sector, aes(x='reorder(sector, rate_per_thousand)', y='rate_per_thousand')) +
 geom_col(color="red", fill="red") +
 geom_text(
     aes(label='rate_per_thousand.round(1)'),
     ha="left",
     nudge_y=.5,
     size=10
 ) +
 labs(
     title="Which Sectors in San Francisco Were Hit the Hardest?",
     x="",
     y="\nClosures per 1,000 Businesses Locations",
     fontstyle="bold",
     nudge_y=.1
 ) +
 coord_flip() +
 theme(
     panel_background = element_blank(),
     axis_ticks_major_x=element_blank(),
     axis_ticks_major_y=element_text(color="grey"),
     plot_title=element_text(family="monospace", face="bold"),
     axis_title_y=element_text(family="monospace"),
     axis_text=element_text(family="monospace"),
     figure_size=[10,7]
 ))
```

#### Closures by Neighborhood

In a similar fashion, we can see how COVID impacted different areas of San Francisco by analyzing the `neighborhoods_analysis_boundaries` metric provided in the data.

```python
sf_records.neighborhoods_analysis_boundaries.count()
```

Just to check, we see that all but a few hundred records we're using have an associated neighborhood. This is great. Let's dive in to see if we can uncover some patterns. We'll start by seeing which neighborhoods are home to the most businesses.

```python
closures_by_neighborhood = (sf_records.
                            groupby(["neighborhoods_analysis_boundaries", "status"])["uniqueid"].
                            count().
                            reset_index().
                            rename(columns={"neighborhoods_analysis_boundaries": "neighborhood"}).
                            pivot_table(index="neighborhood", columns="status", values="uniqueid", fill_value=0).
                            assign(
                                total = lambda x: (x.closed + x.open),
                                pct_closed = lambda x: x.closed / (x.closed + x.open),
                                rate_per_thousand = lambda x: x.pct_closed * 1000
                            ).
                            reset_index().
                            sort_values(by="rate_per_thousand", ascending=False))

closures_by_neighborhood.sort_values(by="total", ascending=False).head()
```

From the above table, we can the Financial District/South Beach area by far has the most companies, followed by the Mission, "South of Market" and Sunset/Parkside areas. Which neighborhoods had the highest closure rate?
